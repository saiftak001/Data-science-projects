```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
import seaborn as sns
```

```python
df = pd.read_csv('Data/Real-Data/Real_Combine.csv')
```

```python
df.head()
```

## Feature Engineering

```python
#First Step is always to check null values
# Checking for null values

sns.heatmap(df.isnull(), yticklabels=False, cmap='viridis')
```

```python
#Drop rows with NaN values.
df = df.apply (pd.to_numeric, errors='coerce')
df = df.dropna()

df
```

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
sns.pairplot(df)
```

```python
df.corr()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df[df.corr().index].corr(), annot=True, cmap="RdYlGn")
```

## Feature Importance
Feature Importance gives you a score for each feature of your data, the higher the score more important or relevant is the feature towards your output variable.
We are using Extra Tree Regressor to extract top 10 feature for the dataset.
Should be used if having huge number of data set with many features
Should not use with current data set

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,Y)
```

```python
X.head()
```

```python
print(model.feature_importances_)
```

```python
feature_importance = pd.Series(model.feature_importances_, index = X.columns)
feature_importance.nlargest(5).plot(kind = 'barh')
```

```python
sns.distplot(Y)
```

## Train Test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
decision_model = DecisionTreeRegressor(criterion='mse')
decision_model.fit(x_train, y_train)
```

```python
print(decision_model.score(x_train, y_train))
```

```python
print(decision_model.score(x_test, y_test))
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(decision_model, X, Y, cv = 5)
```

```python
score.mean()
```

```python
predicted = decision_model.predict(x_test)
```

```python
sns.distplot(y_test-predicted)
```

```python
plt.scatter(y_test, predicted)
```

## Hyperparameter Tuning in DecisionTree 

```python
from sklearn.model_selection import GridSearchCV
dtree = DecisionTreeRegressor()
```

```python
parameters = {
    'criterion':['mse', 'friedman_mse'],
    'splitter':['best', 'random'],
    'max_depth':[ 3, 4, 5, 6, 8, 10, 12, 15],
    'min_samples_leaf':[ 1,2,3,4,5 ],
    'min_weight_fraction_leaf':[0.1,0.2,0.3,0.4],
    'max_features':["auto","log2","sqrt",None ],
    'max_leaf_nodes':[None,10,20,30,40,50,60,70],
    'presort':[True, False]}
dtree_regressor = GridSearchCV(dtree, parameters, scoring='neg_mean_squared_error',cv=10)
```

```python
def timer(start_time = None):
    if not start_time:
        start_time = datetime.now()
        return start_time
    elif start_time:
        thour, temp_sec = divmod((datetime.now()-start_time).total_seconds(), 3600)
        tmin, tsec = divmod(temp_sec, 60)
        print('\n Time taken: %i hours %i minutes and %s seconds.' % (thour, tmin, round(tsec, 2)))
```

```python
from datetime import datetime

start_time = timer(None)
dtree_regressor.fit(x_train,y_train)
timer(start_time)
```

```python
print(dtree_regressor.best_params_)
print(dtree_regressor.best_score_)
```

```python
prediction = dtree_regressor.predict(x_test)
```

```python
sns.distplot(y_test-prediction)
```

```python
from sklearn import metrics;
print('MAE:', metrics.mean_absolute_error(y_test, predicted))
print('MSE:', metrics.mean_squared_error(y_test, predicted))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predicted)))
```
